# Crecimiento exponencial

Vamos a modelar el crecimiento de una supuesta "colonia bacteriana". 

Esto es el crecimiento sin ningún tipo de restricción, por lo que el sistema solo crece y crece, este tambien es conocido como el modelo Malthusiano, debido a que fue Malthus el que lo implemento en un tratado sobre el análisis de la población el cual lleva por nombre **Ensayo sobre el principio de la poblacion**.
El ejemplo de Robert Malthus es interesante, por que es un módelo matemático bien hecho, pero con interpretaciones erroneas, pues los supuestos estan mal hechos, lo cual derivo en deducciones y predicciones incorrectas, la predicción principal es que iba a haber una gran hambruna por que la población iba a crecer en desmedida y eso nunca pasó.

En nuestro caso el modelar el crecimiento de una supuesta "colonia bacteriana" es ideal, debido a que pensaremos en un experimento controlado en el cual se pasa de 2000 bacterias a 9000 bacterias en 3 días.

Recordemos la deducción para este tipo de problemas.
Suponemos que la población crece de modo proporcional al tiempo


$$
\frac{dN(t)}{dt} = k\cdot N(t)
$$

Donde $k$ es la tasa de crecimiento de la población.

Esto corresponde a una ecuación de variables separables, que se resuelve tal que:
$$
\frac{dN}{N} = k\cdot dt
$$

Integrando ambos miembros
$$
\int(\frac{dN}{N})=\int(k\cdot dt)
$$
De aquí tendremos:
$$
ln(N)+C_0 = k\cdot t +C_1
$$

Despejando la constante C_0
$$
ln(N)= k\cdot t+C_1 - C_0
$$

Donde sin perdida de generalidad podemos tomar $C=C_1-C_0$ y tenemos:

$$
ln(N)= k\cdot t+C
$$
Exponenciamos para encontrar $N$

$$
N(t)=e^{k\cdot t+C}
$$
$$
N(t)=e^{C}e^{k\cdot t}

$$
Si evaluamos $t=0$ lo que tendremos es el nivel de poblacion que tendremos al inicio del experimento

$$
N(0)=N_0
$$
Así tendremos:

$$
N_0 = e^{C}e^{k\cdot 0}
$$
$$
N_0 = e^{C}\cdot 1
$$
$$
N_0 = e^{C}
$$

Lo que nos lleva a la ecuación:
$$

N(t) = N_0\cdot e^{kt}
$$

Para nuestro problema, primeramente vamos a estimar $k$

Los datos con los que contamos son:

$N_0:2000$ bacterias

$N_{3_{días}}= 9000$ bacterias

$t: 3$ días

Con estos datos podemos obtener $r$

$r=\frac{\log{\frac{N_0}{N_f}}}{t_{3_{días}}}$

Sustituyendo datos: 

$r=\frac{\log{\frac{2000}{9000}}}{3}$



```python
import numpy as np
import matplotlib.pyplot as plt


# Parámetros del modelo

N0 = 2000        # población inicial
Nf = 9000        # población final
t_dias = 3      # días

# Estimación de la tasa r
r = (np.log(Nf / N0)) / t_dias

# Tiempo continuo
t = np.linspace(0, 6, 100)

# Modelo exponencial
N = N0 * np.exp(r * t)


# Gráfica

plt.figure()
plt.plot(t, N, label="Modelo exponencial")
plt.scatter([0, t_dias], [N0, Nf], label="Datos", marker='o', color='red')

plt.xlabel("Tiempo (días)")
plt.ylabel("Número de bacterias")
plt.title("Crecimiento bacteriano exponencial")
plt.text(
    0.05, 0.9,
    f"tasa de crecimiento r = {r:.4f} (1/día)",
    transform=plt.gca().transAxes,
    fontsize=12,
    bbox=dict(facecolor='white', alpha=0.7)
)
plt.legend()
plt.grid()



print(f"Tasa de crecimiento estimada r = {r:.4f}")
#plt.savefig("../Images/crecimiento_bacteriano_exponencial.png", dpi=300, bbox_inches='tight')
plt.show()
```

# Modelo exponencial basado en una población real

En las poblaciones de **E. Coli** estas pueden crecer hasta el doble cada cierto tiempo.

El módelo empleado en este caso es uno que duplica su población

$N(t)=N_0\cdot 2^{t/T}$

aquí $t$ es el tiempo transcurrido, mientras $T$ es el tiempo de duplicación, sigue siendo una tasa de crecimiento.


fuentes:
https://theory.labster.com/bacterial-growth/

A partir de aquí los datos que tenemos son:

$T:$ tiempo de duplicación, **20 min para E.Coli**





```python
import numpy as np
import matplotlib.pyplot as plt



t_bact = 60  # minutos
T = 20       # minutos

# Vector de tiempo
t = np.linspace(0, 100, 100)

# Modelo por duplicación
N_dup = N0 * 2**(t / T)

# Valor puntual en t = 60
N_60 = N0 * 2**(t_bact / T)

plt.figure()
plt.plot(t, N_dup, label="Modelo por duplicación")
plt.scatter([0,t_bact], [N0,N_60], color='red', label="t = 60 min")

plt.xlabel("Tiempo (minutos)")
plt.ylabel("Número de bacterias")
plt.title("Crecimiento bacteriano por duplicación")
plt.legend()
plt.grid()

plt.show()

plt.savefig("C:/Users/Deadl/OneDrive - Instituto Politecnico Nacional/Documents/Python Scripts/Git/Curiosidades/Images/crecimiento_bacteriano_duplicacion.png", dpi=300, bbox_inches='tight')
```
