# Atractor de Lorenz

El sistema de Lorentz es un ejemplo de uno de esos sistemas caoticos.
A continuacion se presenta el sistema de ecuaciones y el como se construye el atractor de Lorentz, uno de los atractores que surgen a partir del modelo.

$$
\begin{cases}
\frac{dx}{dt} = s(y - x) \\
\frac{dy}{dt} = rx - y - xz \\
\frac{dz}{dt} = xy - bz
\end{cases}
$$

Este presenta una estructura fractal.

+ Es un ejemplo de un sistema determinista
+ Tiene la caracteristica de que no puedes predecir cosas a largo plazo con este sistema determinista
+ Gran ejemplo del "caos determinista"


```python
import numpy as np
from scipy.integrate import odeint
import plotly.express as px
from IPython.display import display, Math



# Definir la función del sistema de Lorenz
def lorenz(u, t):
    s = 10; r = 28; b = 8/3
  #s (parametro sigma): relación entre velocidad de difusión y la difusión térmica
  #r (parametro rho): diferencia de temperaturas entre dos capas de fluido (modelo laminar, no turbulento)
  #b (parametro beta): relación de la sección de la caja de convección, define la disipación de energía dentro de la sección elegida.

    ux_dot = s * (u[1] - u[0])
    uy_dot = r * u[0] - u[1] - u[0] * u[2]
    uz_dot = u[0] * u[1] - b * u[2]
    return [ux_dot, uy_dot, uz_dot]

# Condiciones iniciales
u0 = [0, 1, 1.05]

# Tiempo
t = np.linspace(0, 100, 10000)

# Resolver el sistema de ecuaciones de Lorenz
u = odeint(lorenz, u0, t)

# Crear figura con Plotly Express
fig = px.line_3d(x=u[:, 0], y=u[:, 1], z=u[:, 2], labels={'x': 'Eje X', 'y': 'Eje Y', 'z': 'Eje Z'}, title='Atractor de Lorenz')
fig.show()
```
